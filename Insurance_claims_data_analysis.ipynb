```python
import pandas as pd

df = pd.read_csv("Insurance data_with_amount.csv")
print(df.info())
print(df.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['claim_amount'], bins=30, kde=True)
plt.show()

sns.boxplot(x='segment', y='claim_amount', data=df)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = df[['customer_age', 'vehicle_age', 'claim_amount']]
y = df['claim_status']  # 0 or 1

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

model = RandomForestClassifier()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```
